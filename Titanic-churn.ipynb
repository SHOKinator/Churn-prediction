#Titanic churn prediction


```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as ex
from sklearn.metrics import f1_score as f1
```

```python
df = pd.read_csv("train_and_test2.csv")
df
```

#Feature engineering


```python
columns = ["Age", "Fare", "Sex", "sibsp", "Pclass", "Embarked", "2urvived", "Parch"]
df2 = df[columns]
df2
```

```python
df2.isnull().sum()
```

```python
df2.dtypes
```

```python
df2.dropna(inplace = True)
```

```python
df2.isnull().sum()
```

```python
scaler = StandardScaler()

scaled_data = ["Age", "Fare"]

df2[scaled_data] = scaler.fit_transform(df2[scaled_data])
df2
```

#Data visualization

```python
ex.pie(df2, names = '2urvived', title = 'Proportion of survived vs not survived', hole = 0.33)
```

```python
ex.pie(df2, names = 'Sex', title = 'Male and Female', hole = 0.33)
```

```python
u_v = df2["Parch"].unique()
u_v
```

```python
X = df2.drop(columns = ['2urvived'])
y = df2['2urvived']
```

#Applying SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_resampled, y_train_resampled = smote.fit_resample(X, y)
```

#Model Development

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1000)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

base_model = DecisionTreeClassifier(max_depth = 1)
adaboost = AdaBoostClassifier(base_model,
                              n_estimators = 50,
                              learning_rate = 1.0,
                              random_state = 777)
adaboost.fit(X_train, y_train)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel = "linear", C = 1.0, random_state = 777)
svm.fit(X_train, y_train)
```

```python
logreg_pred = logreg.predict(X_test)

ada_pred = adaboost.predict(X_test)

svm_pred = svm.predict(X_test)
```

#Evaluation

```python
lr_accuracy = accuracy_score(y_test, logreg_pred)
ada_accuracy = accuracy_score(y_test, ada_pred)
svm_accuracy = accuracy_score(y_test, svm_pred)

print("Logistic Regression accuracy:", lr_accuracy * 100)
print("\nAdaBoost accuracy:", ada_accuracy * 100)
print("\nSVM accuracy:", svm_accuracy * 100)
```

```python
fig = go.Figure(data=[go.Table(header=dict(values=['<b>Model<b>', '<b>F1 Score<b>'],
                                          line_color='darkslategray',
                                          fill_color='whitesmoke',
                                          align=['center','center'],
                                          font=dict(color='black', size=18),
                                          height=40),

                              cells=dict(values=[['<b>Logistic Rgeression<b>', '<b>AdaBoost<b>','<b>SVM<b>'],
                               [np.round(f1(logreg_pred, y_test), 2),
                                np.round(f1(ada_pred, y_test), 2),
                                np.round(f1(svm_pred, y_test), 2)]]))
                     ])

fig.update_layout(title='Model Results')
fig.show()
```

```python
from sklearn.model_selection import cross_val_score
logreg_f1_cross_val_scores = cross_val_score(logreg, X, y, cv = 5, scoring = 'f1')
ada_f1_cross_val_scores = cross_val_score(adaboost, X, y, cv = 5, scoring = 'f1')
svm_f1_cross_val_scores = cross_val_score(svm, X, y, cv = 5, scoring = 'f1')
```

```python
fig = make_subplots(rows = 3, cols = 1, shared_xaxes = True, subplot_titles = ('Logistic Regression Cross Val Scores',
                                                                               'AdaBoost Cross Val Scores',
                                                                               'SVM Cross Val Scores'))
fig.add_trace(
    go.Scatter(x=list(range(0, len(logreg_f1_cross_val_scores))), y = logreg_f1_cross_val_scores, name = 'Logistic Regression', mode = 'lines+markers'),
    row = 1, col = 1
)

fig.add_trace(
    go.Scatter(x = list(range(0, len(ada_f1_cross_val_scores))), y = ada_f1_cross_val_scores, name = 'AdaBoost', mode = 'lines+markers'),
    row = 2, col = 1
)

fig.add_trace(
    go.Scatter(x = list(range(0, len(svm_f1_cross_val_scores))), y = svm_f1_cross_val_scores, name = 'SVM', mode = 'lines+markers'),
    row = 3, col = 1
)

fig.update_layout(height = 700, width = 900, title_text = '5 Cross Validation')
fig.update_yaxes(title_text = 'F1 Score')
fig.update_xaxes(title_text = 'Fold #')
fig.show()
```
